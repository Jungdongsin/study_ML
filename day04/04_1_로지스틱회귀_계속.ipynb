#### 데이터준비

다중분류 (7가지 생선 중에서 무얼까?)

```python
fish = pd.read_csv("https://bit.ly/fish_csv")
fish.tail()
```

```python
fish.info()
```

```python
pd.unique(fish.Species)
```

```python
# 훈련을 시킬 입력데이터(다섯가지 특성)
# 2차원 배열로 변경 - to_numpy()
fish_input = fish[["Weight","Length", "Diagonal", "Height", "Width"]].to_numpy()
```

```python
fish_input.shape
```

```python
# ["Weight","Length", "Diagonal", "Height", "Width"] -> [x0, x1, x2, x3, x4]로 변경
fish_input[:5]
```

```python
# 어종에 대해 예측할 결과값
fish_target = fish["Species"].to_numpy()
fish_target[154:]
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(fish_input,fish_target,random_state=42)
# random_state를 변경하면서 훈련
```

```python
print(train_input.shape)
print(test_input.shape)
```

### 단위값을 표준화

```python
# weight:8~1000, width:5~10 => 단위의 갭 차이가 많이 남 (표준화 필요) --> -2~ 2 정도의 실수로 변형
# (단위값)표준화(단위축소) = (실제값 - mean(평균)) / stdev(표준편차) 
from sklearn.preprocessing import StandardScaler # fit, transform() 묶어 놓은 클래스

ss = StandardScaler() # 생성자
ss.fit(train_input) # x0~x4의 평균, 표준편차 등 구해놓음
```

```python
train_scaled = ss.transform(train_input) # 표준화 식 대입해서 표준화시킴
test_scaled = ss.transform(test_input)
```

```python
train_scaled[:20]
```

#### k-최근접 이웃 분류 알고리즘 분류

```python
from sklearn.neighbors import KNeighborsClassifier

kn = KNeighborsClassifier(n_neighbors=5)
kn.fit(X=train_scaled, y=train_target)
```

```python
print(kn.score(X=train_scaled, y=train_target))
print(kn.score(X=test_scaled, y=test_target))
```

```python
kn.classes_
```

#### KNN 분류로 예측

```python
test_scaled[:5]
#표준화된거라 뭔지 잘모름
```

```python
print(test_input[:5])

print(test_target[:5])

print(kn.predict(test_scaled[:5]))
#예측(80~85%정도 맞는듯)
```

```python
# kn = KNeighborsClassifier(n_neighbors=3)일때
print(kn.classes_)
proba = kn.predict_proba(test_scaled[:5]) # 예측 결과값을 한눈에 파악할 수 있는 함수
print(np.round(proba, decimals=4)) #decimals=4 : 소수점4째자리까지 출력
# 1이면 100%확률
```

#### 로지스틱 회귀

```python
z = np.arange(-5, 5, 0.1)
z
```

```python
phi = 1 / (1 + np.exp(-z)) # 시그모이드(로지스틱) 함수 : 무조건 0에서 1사이 값임

plt.figure(figsize=(10,5))
plt.plot(z,phi)
plt.show()
```

#### 이항 로지스틱 회귀(두개만 분류) Bream | Smelt

```python
from sklearn.linear_model import LogisticRegression # 로지스틱 회귀(분류)

lr = LogisticRegression() # 로지스틱 회귀 클래스
lr.fit(X=train_bream_smelt, y=target_bream_smelt)
```

```python
# 정확도
lr.score(X=train_bream_smelt, y=target_bream_smelt)
```

```python
print(target_bream_smelt[:5]) # 실제 타겟값
print(lr.predict(X=train_bream_smelt[:5])) # 예측 값 확인
```

```python
lr.predict_proba(X=train_bream_smelt[:5])
# predict_proba() : ~이 될 확률 예측 함수
# [bream일 확률, smelt일 확률]
```
